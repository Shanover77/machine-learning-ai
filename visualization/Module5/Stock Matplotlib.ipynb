## Stock Prices and Matplotlib

```python
# Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns
```

```python
# Data

df = pd.read_csv(r'data/stockprices.csv')
df.tail()
```

```python
# Google
google = df[df['Name']=='GOOGL'][['close','date']]
google.tail()
google.shape
```

```python
# Apple
apple = df[df['Name']=='AAPL'][['close','date']]
apple.tail()
apple.shape
```

```python
# Amazon
amazon = df[df['Name']=='AMZN'][['close','date']]
amazon.tail()
amazon.shape
```

```python
# Meta/Facebook
fb = df[df['Name']=='FB'][['close','date']]
fb.tail()
fb.shape
```

```python
# Microsoft
ms = df[df['Name']=='MSFT'][['close','date']]
ms.tail()
ms.shape
```
