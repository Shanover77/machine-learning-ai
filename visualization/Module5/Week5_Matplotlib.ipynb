### Line Graph:

Syntax: plt.plot([x], y, [format_string])  
x = optional, default value (0...... n-1)

***Format String Docs:***
    Line Style:
        '-' (Hyphen) - Solid line (default).
        '--' (Double hyphen) - Dashed line.
        '-.' (Dash dot) - Dash-dot line.
        ':' (Colon) - Dotted line.

    Marker Style:
        '.' (Period) - Point marker.
        ',' (Comma) - Pixel marker.
        'o' (Lowercase 'o') - Circle marker.
        'v' (Lowercase 'v') - Downward-pointing triangle marker.
        '^' (Caret) - Upward-pointing triangle marker.
        '<' (Less than) - Left-pointing triangle marker.
        '>' (Greater than) - Right-pointing triangle marker.
        's' (Lowercase 's') - Square marker.
        'p' (Lowercase 'p') - Pentagon marker.
        '*' (Asterisk) - Star marker.
        'h' (Lowercase 'h') - Hexagon1 marker.
        'H' (Uppercase 'H') - Hexagon2 marker.
        '+' (Plus) - Plus marker.
        'x' (Lowercase 'x') - Cross (x) marker.
        'D' (Uppercase 'D') - Diamond marker.

    Color:
        'b' (Blue)
        'g' (Green)
        'r' (Red)
        'c' (Cyan)
        'm' (Magenta)
        'y' (Yellow)
        'k' (Black)
        'w' (White)

```python
# Simple example

plt.plot([2,4,5,6,8], 'bo-') # without an x-values
```

```python
plt.plot([2,4,6,8,10],[2,4,5,6,8], 'o', color='#00aaff', linestyle=':') # with an x-values, changing style
```

*** Multiple data-pairs ***

```python
plt.plot([2,4,6,8],'bs:', [1,4,7,8], 'g^-')
```

**Customize markers and line sizes**  
  
markersize - to set marker's sizes  
linewidth - **  
markerfacecolor - to set a different marker color # in case the line-width is too large to see the markers  

```python
plt.plot([2,4,6,8],'ro--',linewidth = 10, markersize = 18, markerfacecolor = 'white')
```

### Creating figures using Matplotlib

* Create a new figure: ` plt.figure() `
* Plot data-points in form of lines/markers/colors: ` plt.plot() `
* Display created figure : ` plt.show() `
* Save the figure : ` plt.savefig('filename.png') `

```python
# Objective: plot data-points, and save the figure as png

plt.figure(figsize= (4,3))
plt.plot([1,2,4,5], [1,3,4,3], 'ro-')

plt.savefig('redplot.png')
plt.show()
```

# Excercise

Plot a line-graph :
* (1,1)

* (4,4)

Modify the line to have a circular marker and denote these datapoints

```python
x = [1,2,4,5]
y = [1,3,4,3]

plt.figure(figsize=(10,8))
plt.plot(x,y,'ro:', linewidth=6, markersize=10)
plt.savefig('example1.png')
plt.show()
```

### Adding legends and text to the figure

```python
plt.figure()
plt.plot([1,2,3],'mo--',label = "USA")
plt.plot([2,4,6],'g*-',label='Germany')
plt.grid() # grid
plt.legend(loc = "upper right", fontsize= 8)
plt.xlabel('X axis')
plt.ylabel('Y axis', rotation=0, labelpad = 20)
plt.show()
```
