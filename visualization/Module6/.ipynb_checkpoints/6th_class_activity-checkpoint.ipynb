```python
# Import Necessary Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load the data
df = pd.read_csv("CustomersTransactions2021.csv")
df.head()
```

```python
# Sorting the value
df.sort_values("id", inplace=True)
```

```python
# Line Graph
plt.figure(figsize=(15, 8))
plt.plot("id", "txn_total", data=df, label="Transaction Amount")
plt.xlabel("Transaction ID")
plt.ylabel("Transaction Amount in $")
plt.title("Line Graph of Transaction Amount")
plt.legend()
```

```python
# Annotate every marker

plt.figure(figsize=(15, 8))
plt.plot(df["id"], df["txn_total"], "mo:")
plt.xlabel("Transaction ID")
plt.ylabel("Transaction Amount in $")
plt.title("Line Graph of Transaction Amount")
plt.legend()

for x, y in zip(df["id"], df["txn_total"]):
    label = "{:.2f}".format(y)
    plt.annotate(
        label,  # this is the text
        (x, y),  # this is the point to label
        textcoords="offset points",  # how to position the text
        xytext=(0, 10),  # distance from text to points (x,y)
        ha="center",
    )  # horizontal alignment can be left, right or center
```

```python
# Plot in the histogram

plt.hist(df["txn_total"])
```

```python
# Plot in the histogram with 6 bins

plt.hist(df["txn_total"], bins=6)
```

```python
n, bins, patches = plt.hist(df["txn_total"], label="Transaction Amount")
plt.xlabel("Transaction Amount in $")
plt.ylabel("Frequency")
plt.title("Histogram of Transaction Amount")
plt.legend()
```
