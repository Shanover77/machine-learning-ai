```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

### Multi-category bar graph

How to position the bars and ticks so as to center the plot

```python
# Syntax: 
# plt.bar(xaxis_categories, bar_heights, width_per_bar)

courses = ["CSE","Mech","ECE","Aero","Biomed"]
```

```python
boyscount_barheight = [160, 223, 70, 262, 301]
girlscount_barheight = [140, 203, 50, 242, 281]

width_per_bar = w = 0.3 # say

bar1 = np.arange(len(courses)) # centered at 0,1,2,3,4
bar2 = [(i + w) for i in bar1]
```

```python
plt.bar(bar1,boyscount_barheight, width_per_bar, label = "Boys") # boys bar
plt.bar(bar2,girlscount_barheight, width_per_bar, label = "Girls") # girls bar

plt.xlabel('Courses')
plt.ylabel('Number of students in each course')
plt.title('Courses vs #Nos Students')

plt.xticks(bar1 + w/2, courses)
plt.legend()
```
