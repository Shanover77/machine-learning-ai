## 2. IQR - Inter Quartile Range

* Inter quartile range is measure of statistical dispersion, is defined as the difference between the 75th and 25th percentiles of the data.

```python
percentile25 = agDf['Admission grade'].describe()['25%']
percentile75 = agDf['Admission grade'].describe()['75%']

# method -1 : Using describe() to get 25% and 75%
iqr = percentile75 - percentile25
print('IQR - using describe():', iqr)

# method - 2 : Pandas .quartile() to get the 25% and 75%
quartiles = df['Admission grade'].quantile([0.25, 0.75])
iqr = quartiles[0.75] - quartiles[0.25]
print('IQR - using quartile():', iqr)
```
